```python
import pandas as pd                        # It is used for efficient data manipulation and analysis
import numpy as np                         # It is used for numerical computations 
import matplotlib.pyplot as plt            # It is used for visualization of data
import seaborn as sns                      # Advanced visualization library
from datetime import datetime              # For proper date-time format in the given dataset
```

```python
# To read a csv file 
data = pd.read_csv(r'C:\Users\divya\Downloads\householdtask3.csv')
```

```python
# To display the top 10 rows of the dataset
display(data.head(10))
# To display bottom 10 rows of the dataset
display(data.tail(10))
```

```python
# To plot a scatter graph between the columns year and own
plt.scatter(data['year'],data['own'])          # It is used to create x and y axis and plot graph between them
plt.title("Scatter Plot showing relation between YEAR and OWN ")  # This function entitles the graph
plt.xlabel('YEAR')         # It is used to label x axis
plt.ylabel('OWN')          # It is used to label y axis
plt.show()                 # It is used to represent the graph
```

```python
# To plot a line graph between the columns year and own
plt.plot(data['year'])          # To plot a line graph of year column
plt.plot(data['own'])           # To plot a line graph of own column
plt.title("Line Graph of YEAR and OWN column")   # It labels the heading of the graph
plt.xlabel('YEAR')              # To add the label to x axis
plt.ylabel('OWN')               # To add the label to y axis 
plt.show()                      # It represent the detailed graph
```

```python
# To plot a bar graph between columns Year and Own
plt.bar(data['year'], data['own'])   # To plot a bar graph on the dataset named data among columns YEAR and OWN
plt.title("Bar graph between Year and Own")   # This function goves the title to this graph
plt.xlabel('YEAR')     # It labels the x axis of the graph
plt.ylabel ('OWN')     # It labels the y axis of the graph
plt.show()             # It displays the bar chart with details
```

```python
# To plot a histogram on the income column
plt.hist(data['income'])     # It plots the histogram on income column
plt.title("Histogram represents Income")       # It titles the histogram
plt.show()     # It shows the graph with details
```

```python
# To plot a histogram on the Expenditure  column
plt.hist(data['expenditure'])     # It plots the histogram on Expenditure column
plt.title("Histogram represents Expenditure")       # It titles the histogram
plt.show()     # It shows the graph with details
```
