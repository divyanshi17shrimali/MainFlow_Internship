```python
# General question to identify relationship and gather data
questions = ["1. How many people have heart disease and how many people doesn't have heart disease?",
"2. People of which sex has most heart disease?",
"3 People of which sex has which type of chest pain most?",
"4. People with which chest pain are most pron to have heart disease?"]
questions
```

```python
df.target.value_counts()  # It counts the unique value in the target column
```

```python
# plotting bar chart.
df.target.value_counts().plot(kind = 'bar', color=["skyblue", "lightgreen"])
plt.title("Heart Disease values")
plt.xlabel("1 Heart Disease, 0 No heart Disease") 
plt.ylabel("Amount")
plt.xticks(rotation=0)
```

```python
# Plotting Pie Chart
df.target.value_counts().plot(kind = 'pie', figsize = (8, 6)) 
plt.legend(["Disease", "No disease"])
plt.title('Distribution of Heart Disease')
```

```python
df.sex.value_counts() # Count the type of genders
```

```python
# Plotting Pie Chart
df.sex.value_counts().plot(kind = 'pie', figsize = (8, 6))
plt.title('Male Female ratio')
plt.legend(['Male', 'Female']);
```

```python
# Mapping between the column 'target' and 'sex'
pd.crosstab(df.target, df.sex)
```

```python
# Plotting a  histogram
sns.countplot(x = 'target', data = df, hue= 'sex')
plt.xlabel('0 = No heart Disease, 1 = Heart Disease')
plt.title("Heart Disease Frequency for Sex")
```

```python
# In our dataset, there are more than twice as many males as females.

# More than 45% of males in the dataset have heart disease, while more than 75% of females have heart disease.
```

```python
# People of which sex has which type of chest pain most?"
# counting values for different chest pain
df.cp.value_counts()
```

```python
# plotting a bar chart
df.cp.value_counts().plot(kind = 'bar', color = ['brown', 'lightskyblue', 'springgreen', 'khaki'])
plt.title('Chest pain type count')
plt.xticks(rotation=0)
```

```python
pd.crosstab(df.sex, df.cp)  # Mapping of 'sex' and 'cp' columns
```

```python
# Plotting a bar chart to show the distribution of chest pain types by sex
pd.crosstab(df.sex, df.cp).plot(kind= 'bar', color = ['coral', 'lightskyblue', 'plum', 'khaki']) 
plt.title('Type of chest pain for sex')
plt.xlabel('0 = Female, 1 = Male')
plt.xticks(rotation=0)
```

```python
# Most of male has type e chest pain and Least of Male has type 4 pain.
# In case of Female `type` and `type 2 percentage is almost same.
```

```python
#  People with which chest pain are most pron to have heart disease?'
pd.crosstab(df.cp, df.target)
```

```python
# plotting a histogram
sns.countplot(x = 'cp', data = df, hue = 'target')
plt.title("Chest Pain Type Distribution by Heart Disease")  
```
