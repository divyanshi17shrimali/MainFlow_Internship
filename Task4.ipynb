```python
# Importing libraries in the program to perform various operations 
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
```

```python
data = pd.read_csv(r"C:\Users\divya\Downloads\USvideos.csv")  # To read the given data 
```

```python
data.head()   # It represents first 5 rows of the given data frame.
```

```python
data.shape   # It returns the total number of rows and columns of the data set.
```

```python
data = data.drop_duplicates()            # It removes the duplicate data from the dataset.
data.shape                               # It provides the number of rows and columns after the removal of duplicate data from the dataset.
```

```python
data.describe()     # It generate descriptive statistics for each numerical column in the dataset.
```

```python
data.info()    # It describes the basic structure and the metadata of the dataset.
```

```python
coulumns_to_remove = ['thumbnail_link', 'description']
data = data.drop(columns = coulumns_to_remove)     # It will remove the multiple columns which are not required.
data.info()                                        # It represent the updated information about the data.
```

```python
import datetime
# It converts the data in date time format   
data["trending_date"] = data["trending_date"].apply(lambda x: datetime.datetime.strptime(x, '%y.%d.%m'))
data.head(3)   # It represents the updated 3 rows 
```

```python
data['publish_time'] = pd.to_datetime(data['publish_time'])   # It updates the data in date time format 
data.head(2)                                           # It represents the first two rows of dataset
```

```python
# It extracts the month, day, and hour from the 'publish_time' column into new columns 
data['publish_month'] = data['publish_time'].dt.month
data['publish_day'] = data['publish_time'].dt.day
data['publish_hour'] = data['publish_time'].dt.hour
data.head(2)            # It displays the first 2 rows.
```

```python
# It prints and sorts unique values from the "category_id" column of DataFrame
print(sorted(data["category_id"].unique()))
[1, 2, 10, 15, 17, 20, 22, 23, 24, 25, 26, 27, 28, 29, 30, 43]
```

```python
# It creates the category_name column with the human readable names and assigns it corresponding to the category_id
data['category_name'] = np.nan
data.loc[data["category_id"] == 1, "category_name"] = 'Film and Animation'
data.loc[data["category_id"] == 2, "category_name"] = 'Autos and Vehicles'
data.loc[data["category_id"] == 10, "category_name"] = 'Music'
data.loc[data["category_id"] == 15, "category_name"] = 'Pets and Animals'
data.loc[data["category_id"] == 17, "category_name"] = 'Sports'
data.loc[data["category_id"] == 19, "category_name"] = 'Travel and Events'
data.loc[data["category_id"] == 20, "category_name"] = 'Gaming'
data.loc[data["category_id"] == 22, "category_name"] = 'People and Blogs'
data.loc[data["category_id"] == 23, "category_name"] = 'Comedy'
data.loc[data["category_id"] == 24, "category_name"] = 'Entertainment'
data.loc[data["category_id"] == 25, "category_name"] = 'News and Politics'
data.loc[data["category_id"] == 26, "category_name"] = 'How to and Style'
data.loc[data["category_id"] == 27, "category_name"] = 'Education'
data.loc[data["category_id"] == 28, "category_name"] = 'Science and Technology'
data.loc[data["category_id"] == 29, "category_name"] = 'Non Profits and Activism'
data.loc[data["category_id"] == 30, "category_name"] = 'Movies'
data.loc[data["category_id"] == 43, "category_name"] = 'Shows'

data.head()      # It represents the first five rows of data
```

```python
# It creates a bar chart depicting total videos published per year
data['year'] = data['publish_time'].dt.year
yearly_counts = data.groupby('year')['video_id'].count()
yearly_counts.plot(kind='bar', xlabel='Year', ylabel='Total Publish Count', title='Total Publish Video Per Year')
plt.show() 
```

```python
# It plots a bar chart displaying total views per year
data['year'] = data['publish_time'].dt.year
yearly_views = data.groupby('year')['views'].sum()
yearly_views.plot(kind='bar', xlabel='Year', ylabel='Total views', title='Total Views per Year')
plt.xticks(rotation=8)
plt.tight_layout()
plt.show()
```

```python
# It plots a bar chart displaying the total views for the top 5 categories based on 'views'

category_views = data.groupby('category_name')['views'].sum().reset_index()
top_categories = category_views.sort_values(by='views', ascending=False).head(5)

plt.bar(top_categories.index, top_categories['views'])
plt.xlabel('Category Name', fontsize=12)
plt.ylabel('Total Views', fontsize=12)
plt.title('Top 5 Categories', fontsize=15)
plt.xticks(ticks=top_categories.index, labels=top_categories['category_name'], rotation=45, ha='right')  # It sets the x axis position at the angle of 45 degree for a clear view
plt.tight_layout()
plt.show()
```

```python
# It plots a bar chart displaying the video count with respect to the different categories
plt.figure(figsize=(12, 6))
sns.countplot(x='category_name', data=data, order=data['category_name'].value_counts().index, palette='Set3') # The palette function is used to provide different colors to the graph
plt.xticks(rotation=90)
plt.yticks(rotation=20)
plt.title("Video Count by Category")
plt.show()
```

```python
# It plots a bar chart displaying the number of videos published per hour 
videos_per_hour = data['publish_hour'].value_counts().sort_index()
plt.figure(figsize=(12, 6))
sns.barplot(x=videos_per_hour.index, y=videos_per_hour.values, palette="rocket")
plt.title('Number of Videos Published per Hour')
plt.xlabel('Hour of Day')
plt.ylabel('Number of Videos')
plt.show()
```

```python
# It plots a line graph which converts publish time to the date time format and displays the videos published over a time
data['publish_time'] = pd.to_datetime(data['publish_time'])
data['publish_date'] = data['publish_time'].dt.date
video_count_by_date = data.groupby('publish_date').size()
plt.figure(figsize=(12, 6))
sns.lineplot(data=video_count_by_date)
plt.title('Videos Published Over Time')
plt.xlabel('Publish Date')
plt.ylabel('Number of Videos')
plt.xticks(rotation=45)
plt.show()
```

```python
# It creates a scatter plot showing the views and likes on the videos
sns.scatterplot(data=data, x='views', y='likes')
plt.title('Views vs Likes')
plt.xlabel('Views')
plt.ylabel('Likes')
plt.show()
```

```python
# A 2x2 grid of count plot is created to visualize the frequency of various video status flags (comments_disabled, ratings_disabled, video_error_or_removed) in DataFrame data
plt.figure(figsize=(14, 8))
plt.subplots_adjust(wspace=0.2, hspace=0.4, top=0.9)

plt.subplot(2, 2, 1)
g = sns.countplot(x='comments_disabled', data=data)
g.set_title("Comments Disabled", fontsize=16)

plt.subplot(2, 2, 2)
g1 = sns.countplot(x='ratings_disabled', data=data)
g1.set_title("Rating Disabled", fontsize=16)

plt.subplot(2, 2, 3)
g2 = sns.countplot(x='video_error_or_removed', data=data)
g2.set_title("Video Error or Removed", fontsize=16)

plt.show()
```
